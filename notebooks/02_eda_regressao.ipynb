# Análise Exploratória de Distribuição de Erros Logísticos
**Objetivo:** Compreender a dispersão da variável contínua `dias_desvio_entrega` para justificar a escolha de funções de perda baseadas em MAE ou RMSE.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

orders = pd.read_csv('../datasets/olist_orders_dataset.csv')
orders['order_delivered_customer_date'] = pd.to_datetime(orders['order_delivered_customer_date'])
orders['order_estimated_delivery_date'] = pd.to_datetime(orders['order_estimated_delivery_date'])
orders['dias_desvio_entrega'] = (orders['order_delivered_customer_date'] - orders['order_estimated_delivery_date']).dt.days

plt.figure(figsize=(10, 5))
sns.histplot(orders['dias_desvio_entrega'].dropna(), bins=100, color='darkred', kde=True)
plt.axvline(0, color='black', linestyle='--', label='No Prazo Limite (Dia Prometido)')
plt.xlim(-30, 30)
plt.title('Distribuição Frequencial dos Desvios de Prazo na Olist')
plt.xlabel('Dias de Diferença (Valores Positivos = Atraso Real)')
plt.ylabel('Volume de Pedidos')
plt.legend()
plt.show()
```
